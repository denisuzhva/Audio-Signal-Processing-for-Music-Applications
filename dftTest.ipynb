```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Define num of samples and freq
N = 64
k0 = 7
x = np.exp(1j * 2 * np.pi * k0 / N * np.arange(N))
plt.plot(np.real(x))
plt.show()
```

```python
## DFT
X = np.array([])
for k in range(N):
    s = np.exp(- 1j * 2 * np.pi * k / N * np.arange(N))
    X = np.append(X, np.sum(x*s))

plt.plot(np.real(X))
plt.show()
```

```python
## IDFT
y = np.array([])
for n in range(N):
    s = np.exp(1j * 2 * np.pi * n / N * np.arange(N))
    y = np.append(y, np.sum(X*s/N))

plt.plot(np.real(y))
plt.show()
```
