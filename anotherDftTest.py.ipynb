```python
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt
```

```python
f0 = 800
f = 13
N = 800
print(N)
time_series = np.linspace(0, N / f0, N)
signal = np.sin(f * 2.0 * np.pi * time_series)
plt.plot(signal)
plt.show()
```

```python
y = fft(signal)
plt.plot(np.abs(y)[0:20])
```
